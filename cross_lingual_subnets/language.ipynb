```python
import itertools

langs = ["ara", "deu", "eng", "spa", "fas", "fra", "hin", "rus", "swe", "urd", "zho"]
pair_order_list = itertools.permutations(langs, 2)
lang_pairs = list(pair_order_list)
```

```python
seen = []
distances = []
for lang1, lang2 in lang_pairs:
    if (lang1, lang2) in seen or (lang2, lang1) in seen:
        continue

    print(f"{lang1} - {lang2}")
    try:
        dist = l2v.distance("syntactic", lang1, lang2)
    except:
        print("???")
        continue

    seen.append((lang1, lang2))
    distances.append((lang1, lang2, dist))
    print(dist)
```

```python
sorted_distances = sorted(distances, key=lambda tup: tup[2], reverse=True)
```

```python
code_to_lang = {
    "ara": "arabic",
    "deu": "german",
    "eng": "english",
    "spa": "spanish",
    "fas": "farsi",
    "fra": "french",
    "hin": "hindi",
    "rus": "russian",
    "swe": "swedish",
    "urd": "urdu",
    "zho": "chinese",
}
```

```python
for lang1, lang2, dist in sorted_distances:
    print(code_to_lang[lang1], code_to_lang[lang2], dist)
```
