```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the JSON data from the file
with open('cross_eval.json', 'r') as file:
    data = json.load(file)

# Convert JSON data to Pandas DataFrame
df = pd.DataFrame(data)

# Create the heatmap
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(df, annot=True, cmap='viridis', linewidths=0.5)

# Add title and labels
plt.title('Evaluation Scores Heatmap')
plt.xlabel('Evaluation Language')
plt.ylabel('Pruned Language')

# Display the heatmap
plt.show()
```
